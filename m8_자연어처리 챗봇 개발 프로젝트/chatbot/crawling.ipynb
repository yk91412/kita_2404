```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

options = Options()
options.add_argument("--window-size=800,600")
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

url = 'https://new-m.pay.naver.com/savings/list/parking'
driver.get(url)

time.sleep(2)  # 페이지 로딩 대기

data = []

try:
    # 첫 번째 요소부터 시작하도록 수정
    elements = driver.find_elements(By.CSS_SELECTOR, '.ProductList_link__pMmxO')
    print(f"Found {len(elements)} elements")  # 요소 개수 확인

    for index in range(len(elements)):
        elements = driver.find_elements(By.CSS_SELECTOR, '.ProductList_link__pMmxO')
        driver.execute_script("arguments[0].scrollIntoView(true);", elements[index])
        time.sleep(1)
        driver.execute_script("arguments[0].click();", elements[index])

        # 데이터 추출
        try:
            title = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/div[1]/div[1]/h3'))
            ).text
        except Exception as e:
            title = None
            print(f"Title extraction failed: {e}")

        try:
            bank = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="content"]/div[1]/div[1]/span[2]'))
            ).text
        except Exception as e:
            bank = None
            print(f"Bank extraction failed: {e}")

        try:
            tab_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CLASS_NAME, 'TabPanelButton_button__JjUkq'))
            )
            tab_button.click()
            time.sleep(1)
        except Exception as e:
            tab_button = None
            print(f"Tab button click failed: {e}")

        try:
            payment_label = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="PRODUCT_GUIDE"]/dl/div[5]/dt'))
            ).text
        except Exception as e:
            payment_label = None
            print(f"Payment label extraction failed: {e}")

        info_dict = {
            "금액": None,
            "가입방법": None,
            "가입대상": None,
            "우대조건": None,
            "이자지급": None
        }

        if payment_label == "이자지급":
            for i in range(1, 6):
                xpath = f'//*[@id="PRODUCT_GUIDE"]/dl/div[{i}]/dd'
                try:
                    text = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.XPATH, xpath))
                    ).text
                    if i == 1:
                        info_dict["금액"] = text
                    elif i == 2:
                        info_dict["가입방법"] = text
                    elif i == 3:
                        info_dict["가입대상"] = text
                    elif i == 4:
                        info_dict["우대조건"] = text
                    elif i == 5:
                        info_dict["이자지급"] = text
                except Exception as e:
                    print(f"Info {i} extraction failed: {e}")
                    continue
        else:
            for i in range(1, 5):
                xpath = f'//*[@id="PRODUCT_GUIDE"]/dl/div[{i}]/dd'
                try:
                    text = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.XPATH, xpath))
                    ).text
                    if i == 1:
                        info_dict["금액"] = text
                    elif i == 2:
                        info_dict["가입방법"] = text
                    elif i == 3:
                        info_dict["가입대상"] = text
                    elif i == 4:
                        info_dict["이자지급"] = text
                except Exception as e:
                    print(f"Info {i} extraction failed: {e}")
                    continue

        try:
            interest_table = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="INTEREST_RATE_GUIDE"]/div[1]/div/div/table'))
            )
            
            thead = interest_table.find_element(By.TAG_NAME, "thead")
            headers = [header.text.strip() for header in thead.find_elements(By.TAG_NAME, "th")]

            tbody = interest_table.find_element(By.TAG_NAME, "tbody")
            rows = tbody.find_elements(By.TAG_NAME, "tr")

            base_interest_rate_dict = {}
            for row in rows:
                th_elements = row.find_elements(By.TAG_NAME, "th")
                td_elements = row.find_elements(By.TAG_NAME, "td")

                if th_elements and td_elements:
                    th_texts = [th.text.strip() for th in th_elements]
                    td_texts = [td.text.strip() for td in td_elements]

                    for th_text, td_text in zip(th_texts, td_texts):
                        base_interest_rate_dict[th_text] = td_text
            base_interest_rate = base_interest_rate_dict
        except Exception as e:
            print(f"Table extraction failed: {e}")
            try:
                base_interest_rate = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, '//*[@id="INTEREST_RATE_GUIDE"]/div[1]/dl/div/dd'))
                ).text
            except Exception as e:
                base_interest_rate = None
                print(f"Fallback extraction for base interest rate failed: {e}")

        try:
            additional_info = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="INTEREST_RATE_GUIDE"]/div[2]/dl/div[1]/dt'))
            ).text
        except Exception as e:
            additional_info = None
            print(f"Additional info extraction failed: {e}")

        try:
            additional_text = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="INTEREST_RATE_GUIDE"]/div[2]/dl/div/dd'))
            ).text
        except Exception as e:
            additional_text = None
            print(f"Additional text extraction failed: {e}")

        condition_text = additional_text if additional_info == "조건별" else None

        data.append({
            "title": title,
            "bank": bank,
            "금액": info_dict.get("금액"),
            "가입방법": info_dict.get("가입방법"),
            "가입대상": info_dict.get("가입대상"),
            "우대조건": info_dict.get("우대조건"),
            "이자지급": info_dict.get("이자지급"),
            "기본이자율": base_interest_rate,
            "조건별": condition_text,
            "유형": additional_text
        })

        driver.back()
        time.sleep(2)

except Exception as e:
    print("Error:", e)

driver.quit()

# DataFrame으로 변환
df = pd.DataFrame(data)
print(df.head())
```
