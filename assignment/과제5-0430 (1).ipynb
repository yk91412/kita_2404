```python
# Task1_0430. 남녀 파트너 정해주기 프로그램(zip)

# 같은 수의 남녀 모임에서 파트너를 랜덤하게 정해주는 프로그램을 만들어 보세요

# random.choice(sequence): 주어진 시퀀스(리스트, 튜플 등)에서 무작위로 아이템을 선택하여 반환.
# random.sample(population, k): 주어진 시퀀스에서 무작위로 k개의 항목을 뽑아 새 리스트로 반환.
male = ['철수','갑돌','로미오','이몽룡','온돌']

female = ['미미','갑순','줄리엣','성춘향','평강']
# male에서 하나 female에서 하나씩을 묶어서 zip으로
import random

# 내가 푼 답안
# 리스트의 길이만큼 랜덤으로 뽑아 리스트 형식으로 변수에 저장하고 zip을 이용해 인덱스 별로 파트너를 정했다
random_male = random.sample(male,len(male))
random_female = random.sample(female,len(female))
for male1,female1 in zip(random_male,random_female):
    print(f"{male1}와/과 {female1}은/는 파트너입니다")
```

```python
# Task5_0430. 사용자로부터 숫자를 입력받아 해당 숫자의 팩토리얼을 계산하세요.

while 1:
    try :
        result = 1
        num = int(input("정수를 입력하세요 : "))
        for i in range(num,0,-1):
            result *= i
        print(result)
        break
    except ValueError:
        print("정수를 입력하세요")
```

```python
# 곱하기 형태도 보여주기
while 1:
    try :
        result = 1
        num = int(input("정수를 입력하세요 : "))
        for i in range(num,0,-1):
            if i > 1:
                print(f"{i} X", end = ' ')
            else :
                print(f"{i} =", end = ' ')
            result *= i
        print(result)
        break
    except ValueError:
        print("정수를 입력하세요")
```

```python
# Task10_0430. 아래 사항을 반영하여 커피 자판기 프로그램을 작성하세요.
# - 시나리오 : 자판기 커피 재고 5잔, 커피 1잔 가격 300원, 재고 범위내에서 300원 이상 돈을 넣으면 거스름돈과 커피를 주고 그렇지 않으면 반환하며 재고가 소진되면 안내멘트 출력
# - 각 Case별 멘트 출력은 상황에 맞게 창작
# - while, if ~ elif ~ else 제어문을 사용하여 작성

# 돈이 300원 안될 때 돈이 부족합니다
# 300~ 1500원 사이일 때 1. 거스름돈이 없다면 몫만큼 커피를 준다 2. 몫만큼 커피를 주고 전체에서 몫을 뺀만큼 거스름돈을 준다
# 그러나 여러번 반복해서 5잔 이상이 되었다면 재고가 소진되었습니다
# 총 커피수를 정해서 + 시켜야함
# 1500원 이상일 때 재고가 소진되었고 재고내의 커피수만 입력

while 1:
    try :
        coffee = 0
        while coffee < 5:
            price = int(input("넣으신 돈을 입력하세요(원 제외) : "))
            if price < 300:
                print("돈이 부족합니다. 커피 한 잔의 값은 300원입니다")
            elif 300<=price<=1500:
                    if price % 300 == 0:
                        print(f"커피 {price//300}잔 드립니다. 거스름돈은 없습니다")
                        coffee += (price//300)
                        if coffee >= 5 :
                            print("판매가 종료되었습니다")
                            break
                        else :
                            pass
                    else :
                        print(f"커피 {price//300}잔 드립니다. 거스름돈은 {price%300}원입니다")
                        coffee += (price//300)
                        if coffee >= 5 :
                            print("판매가 종료되었습니다")
                            break
                        else:
                            pass
            else :
                if price % 300 == 0:
                    print(f"커피 5잔 드립니다. 거스름돈은 없습니다")
                    print("판매가 종료되었습니다")
                    break
                else :
                    print((f"커피 5잔 드립니다. 거스름돈은 {price-1500}원 입니다"))
                    print("판매가 종료되었습니다")
                    break
        break
    except ValueError :
        print("수를 입력하세요")
```
