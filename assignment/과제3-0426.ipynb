## 문제 1번

```python
# Task1_0426. 주어진 리스트에서 중복된 요소를 제외하고 고유한 요소의 개수를 세는 프로그램을 작성하세요.

data = [1, 2, 2, 3, 4, 4, 4, 5]
set1 = set(data)
print(set1)
print(len(set1))
```

```python
# 함수로 푸는 방법
def duplication_except1(data):
    li1 = []
    for i in data:
        if i not in li1 :
            li1.append(i)
    return li1
data = [1, 2, 2, 3, 4, 4, 4, 5]
print(duplication_except1(data))
print(len(duplication_except1(data)))
```

```python
# 연속되는 문자 제거하기
# 리스트의 맨 끝이랑 비교한 이유는 연속되는 문자만 없으면 되기때문에 !
def duplication_except(data):
    li1 = [data[0]]
    for i in data[1:]:
        if i != li1[-1]:
            li1.append(i)
    return li1
print(duplication_except(data))
```

## 문제 2번

```python
# Task2_0426. 두 개의 리스트에서 공통으로 나타나는 요소들을 찾아 리스트로 반환하는 프로그램을 작성하세요.

# 예시 입력: list1 = [1, 2, 3, 4, 5], list2 = [4, 5, 6, 7, 8]

# 예시 출력: [4, 5]

def intersaction_factor(li1,li2):
       s1 = set(li1)
       s2 = set(li2)
       return list(s1 & s2)

li1 = [1,2,3,4,5]
li2 = [4,5,6,7,8]
result = intersaction_factor(li1,li2)
print(result)
```

## 문제 3번

```python
# Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

# - numbers = [1, 2, 3, 10, 20]
# - threshold = 15

# check greater than threshold

def max_number(numbers, threshold):
    for i in numbers:
        if i > threshold :
            # 큰 요소가 하나라도 있으면 되므로 있다면 함수를 종료해도된다
            return True
            # 반복문을 다 돌았는데도 큰 수가 없다면 False를 리턴
    return False

numbers = [1,2,3,10]
threshold = 15
print(max_number(numbers,threshold))
```

```python
# 반환값이 없다면
numbers = [1,2,3,10,20]
threshold = 25
# any : 하나라도 참이라면
if any(num > threshold for num in numbers):
    print("Ture")
else :
    print("False")
```

## 문제 4번

```python
# Task4_0426. 주어진 문자열이 특정 문자열을 포함하는지 확인하고 결과를 불 값으로 반환하는 함수를 작성하세요.

text = "hello world"
substring = "world"

def string_cheak(text):
    text1 = text.split()
    for i in text1 :
        if i == substring :
            return True
    return False

print(string_cheak(text))
```

```python
def contains_substring(text,substring):
    if substring in text :
        return True
    else:
        return False

text = "hello world"
substring = "app"
result = contains_substring(text,substring)
print(result)
```

## 문제 5번

```python
# Task5_0426. 주어진 연도가 윤년인지 판별하는 함수를 작성하세요. 윤년은 다음의 조건을 만족해야 합니다:
# - 4로 나누어 떨어진다.
# - 100으로 나누어 떨어지지 않거나, 400으로 나누어 떨어지면 윤년이다.

# year = 2020

# True
def year_check(year):
    if (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0):
        return True
    else :
        return False

year = 2020
print(year_check(year))
```

```python
def is_leap_year(year):
    # %이 나머지 이기 때문에 not %는 나머지가 없고를 뜻함
    if (not year % 4 )and ((year %100)or (not year % 400)):
        return True
    else :
        return False

year = 2024
is_leap_year(year)
```

## 문제 7번

```python
# Task7_0426. 다음 과제를 수행하세요.
# - 사용자로 부터 텍스트를 입력 받는다.
# - 문자을 단어 단위로 분리 : split()
# - 단어의 빈도수를 저장할 딕셔너리를 생성
# - 각 단어의 빈도 수를 계산(for 문 / if else문)
# - 결과 출력

# [ 예시 ]

# 문장을 입력하세요: I love apple. I love orange. Apple is tasty

# {'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}
text = input("텍스트를 입력하세요 : ")
text_lower = text.lower()
text_divide = text_lower.split()

dic = {}
for i in text_divide:
    if i in dic:
        dic[i] += 1
    else :
        dic[i] = 1
print(dic)
```

```python
# 함수로 변경
def word_frequency2(text):
    text = input("텍스트를 입력하세요 : ")
    text_lower = text.lower()
    text_divide = text_lower.split()
    dic = {}
    for i in text_divide:
        if i in dic:
            dic[i] += 1
        else :
            dic[i] = 1
    print(dic)

word_frequency2(text)
```

```python
def word_frequency(text):
    sentence = input("문장을 입력하세요 : ")
    words = sentence.split()
    word_frequency={}
    for word in words:
        word = word.lower()
        if word in word_frequency:
            word_frequency[word] += 1
        else :
            word_frequency[word] =1

    print(word_frequency)

word_frequency(text)
```
