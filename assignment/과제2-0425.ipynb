### 문제 1

```python
# Task1_0425. 주어진 숫자 리스트에서 최소값과 최대값을 찾아 출력하세요.
# 내가 풀었던 답
numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]
print(max(numbers))
print(min(numbers))
```

```python
# 새 변수에 값을 넣어 f포맷을 활용하여 출력
min_value = min(numbers)
max_value = max(numbers)

print(f"최댓값은 : {max_value}이고, 최솟값은 : {min_value}입니다")
```

```python
# 반복문과 조건문을 활용하여 두 수를 비교후 작은 수는 최솟값으로 큰 수는 최댓값으로 넣는 방법
# 최솟값과 최댓값을 첫번째 요소로 초기화
max_value = numbers[0]
min_value = numbers[0]

for number in numbers:
    if number > max_value:
        max_value = number
    if number < min_value:
        min_value = number
print(f"최댓값은 : {max_value}이고, 최솟값은 : {min_value}입니다")
```

### 문제 5

```python
# Task5_0425 주어진 정수 리스트와 회전 횟수 k에 대해 리스트를 오른쪽으로 k만큼 회전시킨 결과를 반환하세요.
# k가 리스트의 길이보다 클 수 있으며, 이 경우 k를 리스트 길이로 나눈 나머지만큼 실제 회전시키면 됩니다.

# 예를 들어, [1, 2, 3, 4, 5]와 k=2가 주어지면, 결과는 [4, 5, 1, 2, 3]이 되어야 합니다.

# k = len(li1) % k 만큼 회전
# k만큼 뒤에 있는 수가 앞으로 ...

def rotate_list(nums,k):
    if not nums :
        return nums
        # 회전횟수를 리스트 길이로 나눈 나머지로 계산
    k = k % len(nums)
    # 회전횟수만큼 끝부분을 앞으로, 앞부분을 뒤로 순서변경
    rotated = nums[-k:]+nums[:-k]
    return rotated

nums = [1,2,3,4,5]
k = 2
print(rotate_list(nums,k))
```

```python
# 왼쪽으로 k만큼 회전시킨 결과
def rotate_left(nums,k):
    if not nums:
        return nums

    k = k %len(nums)
    rotate_list = nums[k:]+nums[:k]
    return rotate_list

print(rotate_left(nums,k))
```
