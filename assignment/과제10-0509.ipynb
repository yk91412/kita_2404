```python
# 사용자 함수 적용
# Task1_0509. 주어진 텍스트에서 문자 빈도를 아래와 같이 출력하세요.(사용자 함수 사용)
import re

def getTextFreq(filename):
    with open(filename,'r') as f:
        filename = f.read()
    filename = re.sub(r'[,\.\'!\n\\]', '', filename).lower()
    filename = filename.split()

    word_req = {}
    for i in filename:
        for word in i:
            if word in word_req:
                word_req[word] += 1
            else:
                word_req[word] = 1

    word_req = sorted(word_req.items() ,key = lambda x: (x[1],x[0]), reverse = True)
    for word in range(min(5,len(word_req))):
        key, value = word_req[word]
        print(f"가장 많이 나온 문자[{key:^3}] -> [{value:^4}]회 나타났습니다")


getTextFreq('mydata.txt')
```

```python
def textfreq(filename):
    with open(filename,'r') as f:
        text = f.read()
        text = re.sub(r'[ ,.!\n\']', '', text)
        fa = {}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
    return fa

ret = textfreq('mydata.txt')
ret = sorted(ret.items(), key = lambda x: x[1], reverse = True)
for c, freq in ret[:5]:
    if c == '\n':
        continue
    print(f"[{c}] -> [{freq}]회 나타남")
```

Task4_0509. **args를 이용하여 아래와 같이 출력할 수 있는 사용자 함수 func2를 작성하세요.

Width: 100\

```python
def func2(*args,**kwargs):
    print(f"Width : {args[0]}")
    print(f"Height : {args[1]}")
    print(f"Other arguments : {kwargs}")

func2(100,200,color='red',thickness=5)
print()
func2(100,200,color='red',thickness=5, type = 'large')
```

```python
def func3(*args,**kwargs):
    print(f"Width : {args[0]}, Height : {args[1]}", end = ', ')
    print("Color :", kwargs.get('color', 'blue'), ", Thickness :" , kwargs.get('thickness', 1))

func3(100,200)
func3(100,200,color = 'red',thickness = 15)
```

```python
# Task10_0509. 위의 계산을 연산자를 choice로 가변적인 숫자들을 *args로 사용한 사용자 함수 cal을 이용하여 다음을 수행하세요.

# cal('*',1,2,3,4,5)
# cal('+',1,2,3,4,5,6,7)
# cal('-',1,2,3,4,5,6,7)
# cal('/',10,3)-> 0일수도 있으니 에러 처리함께
def cal(choice,*args):
    if choice == '+':
        print(sum(args))
    elif choice == '-':
        total_minus = 0
        for i in args:
            total_minus -= i
        print(total_minus)
    elif choice == '*':
        total_mul = 1
        for i in args:
            total_mul *= i
        print(total_mul)
    elif choice == '/':
        total_divide = args[0]
        for i in args[1:]:
            if i != 0:
                total_divide /= i
                print(total_divide)
            else:
                print("0으로 나눌 수 없습니다")
    else:
        print("연산 할 수 없는 연산자 입니다")

cal('/',10,4)
```

```python
def cal(choice,*args):
    result = args[0]
    for i in args[1:]:
        if choice == '+':
            result += i
        elif choice == '-':
            result -= i
        elif choice == '*':
            result *= i
        elif choice == '/':
            if i == 0:
                return print("오류")
            else:
                result /= i
        else:
            return "계산할 수 있는 기호가 아닙니다"
    return result

cal('/',10,4)
```
