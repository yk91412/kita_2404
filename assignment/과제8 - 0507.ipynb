```python
# Task1_0507. 튜플 형태인 strings를 다양한 문자가 포함된 순서로 정렬하세요.
# 내가 푼 방법 1
strings = ('a', 'ca', 'bar', 'aaaa', 'ababa')

strings_sort = sorted(strings, key = lambda x: len(set(x)),reverse=True)
print(strings_sort)
```

```python
strings = ('a', 'ca', 'bar', 'aaaa', 'ababa')
strings = list(strings)
# 방법2
for index in range(len(strings)):
    strings_len = len(set(strings[index]))
    strings[index] = (strings[index], strings_len)

strings_sort = sorted(strings,key = lambda x : (-x[1]))
print(strings_sort)
```

```python
strings = list(strings)
strings.sort(key = lambda x : len(set(x)), reverse = True)
print(list(strings))
```

```python
# Task2_0507. a,b 변수와 연산자를 입력하면 사칙연산을 수행하는 코드를 작성하세요.(eval 사용하는 경우와 안하는 경우)

def calculate(a,b,express):
    if express == '+':
        print(a+b)
    elif express =='-':
        print(a-b)
    elif express == '*':
        print(a*b)
    elif express == '/':
        print(a/b)
    else :
        print("계산 불가능한 연산자 입니다")

if __name__ == "__main__":
    a = int(input("첫번째 수를 입력하세요 >> "))
    b = int(input("두번째 수를 입력하세요 >>"))
    express = input("원하는 연산자를 입력하세요. 보기 (+ - / *)>> ")
    calculate(a,b,express)
```

```python
def calcu(text):
    print(text)

if __name__ == "__main__":
    text = eval(input("원하는 계산 수식을 입력하세요 : "))
    calcu(text)
```

## 4번

```python
# Task4_0507. 주어진 문자열 리스트의 요소들을 대문자로 변환하세요.
words = ["hello", "world", "python", "map"]

words_upper = [i.upper() for i in words]
print(words_upper)
```

```python
# 인덱스를 이용
words = ["hello", "world", "python", "map"]
for i in range(len(words)):
    words[i] = words[i].upper()
print(words)
```

```python
# str.upper는 문자열의 모든 문자를 대문자로 변환하는 메서드
# 반복문을 돌리지 않고 간단하게 대문자로 변환할 수 있다
words = ["hello", "world", "python", "map"]
upper_words = list(map(str.upper,words))
print(upper_words)
```

```python
# 요소를 이용
result = []
words = ["hello", "world", "python", "map"]
for word in words:
    result.append(word.upper())
print(result)
```

```python
words = ["python", "is", "powerful", "programming", "language", "pandas"]
# 방법2
words_find = [i for i in words if i.find('p') == 0]
print(words_find)
```

find와 index 차이점

포함되지 않은 문자를 찾게되면 index()는 에러를 발생하지만, find()는 -1을 리턴합니다.

그리고 index()는 list에서도 사용이 가능하지만 find는 문자열에서만 사용 가능합니다

```python
# 방법3
words = ["python", "is", "powerful", "programming", "language", "pandas"]
word_find = [i for i in words if i.startswith('p')]
print(word_find)
```
