```python
# 주어진 문자열 리스트에서 유효한 Python 변수 이름만을 추출하여 반환하는 함수를 작성하세요
identifiers = ["var1","2thins","variable_name","time!"]
```

```python
# isidentifier()는 파이썬의 문자열 메서드로, 주어진 문자열이 유효한 식별자인지를 확인
# isidentifier()는 예약어인지는 확인하지 않음
def valid_identifiers(identifiers):
    valid = []
    for identifier in identifiers:
        if identifier.isidentifier():
                valid.append(identifier)
    return valid

identifiers = ["var1","2thins","variable_name","time!","_totay", "whatis_","False"]
print(valid_identifiers(identifiers))
```

```python
import keyword
def valid_identifiers(identifiers):
    valid = []
    for identifier in identifiers:
        if identifier.isidentifier() and not keyword.iskeyword(identifier): #keyword.iskeyword(s) iskeyword는 괄호안에 s라는 값을 넣어야 가능하다
                valid.append(identifier)
    return valid

identifiers = ["var1","2thins","variable_name","time!","_totay", "whatis_","False"]
print(valid_identifiers(identifiers))
```

```python
# 1.알파벳이 맞는지 확인: isalpha
# 2. 숫자가 맞는지 확인   : isdigit,isnumeric
# 3. 공백이 맞는지 확인   : isspace
# 특수문자는 알파벳 숫자 공백을 제외하고

def vaild_name(name):
    for i in name:
        if i[0].isalpha() and (i[-1].isnumeric() or i[-1].isalpha()) :
            print(i)

identifiers = ["var1","2thins","variable_name","time!"]
vaild_name(identifiers)

# 4.24에 풀었던 답안
```

```python
# isalnum 문자열이 숫자 영어 한글로되어있으면 참 or 아니면 거짓
# 문자열 중간에 공백이 있으면 False는 isspace로 불가능...

def vaid_name(name):
    valid = []
    for i in name:
        if (i[0].isalpha() or i[0] == "_") and (i[-1].isalnum() or i[-1] == "_") and not i.isspace():
            valid.append(i)
    return valid

identifiers = ["var1","2thins","variable_name","time!","_totay", "whatis_", "what time"]
print(vaid_name(identifiers))

# 4.25 다시 풀어봤으나 사이에 있는 공백은 제거하지 못함
```

```python
# 제거되는 시점에서 인덱스 번호가 변경됨
for i in identifiers:
    if i[0].isdigit() or not i.isalnum():
        identifiers.remove(i)
print(identifiers)

# 다른분이 풀었던 방식
```
