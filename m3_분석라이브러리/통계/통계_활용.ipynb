Q. 한 학급의 학생 10명의 수학 시험 점수는 다음과 같습니다:

[85, 72, 90, 68, 88, 76, 95, 89, 77, 84].

학급의 평균 점수와 표준편차를 계산하여 전체 성적 수준과 변동성을 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import numpy as np
student = np.array([85, 72, 90, 68, 88, 76, 95, 89, 77, 84])

print(f"평균점수 : {student.mean()}")
print(f"표준편차 : {student.std()}")
```

- 평균 점수에 비해서 흩어져있는 학생들의 점수 확인
- 표준편차의 값에 따라서 학생들의 점수 차가 큰지 작은지 알 수 있다(값이 크다면 점수차이가 큰것)
- 표준편차에 값에 따라서 데이터의 변동성이 크고 작은지 알 수 있다(값이 크다면 변동성이 크다)


```python
# 인사이트 해석
# 평균은 학급의 전체 성적 수준을 나타내고, 표준편차는 성적의 변동성을 나타냅니다
# 도메인 지식을 활용하여 표준편차가 10보다 크다면 변동성이 큰걸로 설정

if student.std() >10:
    print("성적 변동성이 큽니다")
else:
    print("성적 변동성이 작습니다")
```

평균이 82.4으로 학급 전체의 성적 수준이 비교적 높은 편입니다.

표준편차가 8.3으로 성적의 변동성이 크지 않으며 학생들이 성적이 비교적 분포되어 있음을 알 수 있습니다

Q. 한 회사의 직원 연봉 데이터가 다음과 같습니다:

[45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000].

이 데이터를 이용하여 연봉의 분포와 이상치를 시각화하는 상자 그림을 그리고, 연봉 분포에 대한 인사이트를 도출하세요.

```python
import numpy as np
import matplotlib.pyplot as plt

data = [45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000]


plt.boxplot(data)
plt.show()
```

- 연봉 이상치가 없으므로 비슷한 연봉을 가지고 있다
- 공식을 이용하여 이상값을 알아낼 수 있다
IQR = Q3 - Q1

이상값 : Q1 - 1.5 * IQR 보다 작거나 Q3 + 1.5 * IQR 클 때

```python
salaries = np.array([45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000])

plt.boxplot(salaries)
plt.title('Box Plot of Employee Salaries')
plt.ylabel('Salary')
plt.show()

# 인사이트 해석
median_salary = np.median(salaries)
q1 = np.percentile(salaries,25)
q3 = np.percentile(salaries,75)
iqr = q3-q1

print(f"Median Salary : {median_salary}")
print(f"Q1 : {q1}, Q3 : {q3}, IQR : {iqr}")

# 이상치 확인
lower_bound = q1 - 1.5 *iqr
upper_bound = q3 + 1.5*iqr
outliers = salaries[(salaries < lower_bound) | (salaries > upper_bound)]

print(f"Outliers : {outliers}")
```

Q. 한 마케팅 캠페인에서 투자한 광고비와 발생한 매출 데이터가 각각 다음과 같습니다:

광고비 [1, 2, 3, 4, 5] (단위: 천 달러),

매출 [10, 15, 20, 25, 30] (단위: 천 달러).

광고비와 매출 간의 관계를 시각화하기 위해 산점도를 그리고, 두 변수 간의 관계에 대한 인사이트를 도출하세요.

```python
ad = [1000,2000,3000,4000,5000]
sales = [10000,15000,20000,25000,30000]

plt.scatter(ad,sales)
plt.show()
```

- 광고비가 올라갈 때 매출도 오른다(비례관계)

```python
ad = np.array([1,2,3,4,5])
sales = ad+np.random.randint(10,size=5)

plt.scatter(ad,sales)
plt.title('Scatter Plot of Ad Spend vs Sales')
plt.xlabel('Ad Spend (thousand dollars)')
plt.ylabel('Sales (thousand dollars)')
plt.show()

# 인사이트 해석
correlation = np.corrcoef(ad,sales)[0,1]
print(f"Correlation coefficient: {correlation}")

if correlation > 0.7:
    print("광고비와 매출 간에 강한 양의 상관 관계가 있습니다")
else:
    print("광고비와 매출 간에 상관관계가 약하거나 없습니다")
```

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
promotion =  [500, 600, 700, 800, 900]
visitor = [300, 400, 500, 600, 700]

df = pd.DataFrame(promotion,visitor)
df.reset_index(inplace=True)
df.columns=['promotion','visitor']
df.corr()
```

- 상관계수가 1로 완벽한 양의 선형 상관관계를 보이고 있다(비례관계)

인사이트

홍보 비용과 방문자 수 간의 상관 계수가 1로
홍보비용이 증가함에 따라 방문자수가 증가하는 강한 양의 상관관계가 있음을 알 수 있습니다

이는 홍보비용을 늘리면 방문자 수가 늘 수 있다는걸 알 수 있습니다

```python
# 홍보 비용과 방문자 수 데이터
promotion_cost = np.array([500,600,700,800,900])
visitor_count = np.array([300,400,500,600,700])
v_vi = visitor_count[::-1]
# v_vi = np.sort(visitor_count)[::-1]
# 상관계수 계산
correlation = np.corrcoef(promotion_cost, v_vi)[0,1]
print(f"Correlation coefficient: {correlation}")

# 인사이트 해석
if correlation > 0.7:
    print("홍보 비용과 방문자 수 간에 강한 양의 상관관계가 있습니다")
elif correlation < 0.7:
    print("홍보 비용과 방문자 수 간에 강한 음의 상관관계가 있습니다")
else:
    print("홍보 비용과 방문자 수 간에 상관관계가 없거나 적습니다")
```

인사이트

홍보 비용과 방문자 수 간의 상관 계수가 -0.9 로 홍보비용이 증가함에 따라 방문자수가 감소하는 강한 음의 상관관계가 있음을 알 수 있습니다

- 일정한 빈도를 가지고 있다

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


house_prices = np.array([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

# 밀도 그래프 그리기
sns.kdeplot(house_prices, fill=True)
plt.title("Density Plot of House Prices")
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()

# 인사이트 해석
mean_price = np.mean(house_prices)
median_price = np.median(house_prices)

print(f"Mean Price : {mean_price}, Median Price : {median_price}")

if mean_price > median_price:
    print("주택 가격 분포가 오른쪽으로 치우쳐 있음을 의미합니다")
elif mean_price < median_price:
    print("주택 가격 분포가 왼쪽으로 치우져 있음을 의미합니다")
else:
    print("주택 가격이 고르게 분포되어 있음을 의미합니다")
```
