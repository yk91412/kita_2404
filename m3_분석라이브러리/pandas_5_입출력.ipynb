```python
# csv 파일 읽고 저장하기
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))
df.to_csv('test.csv')
# df.to_csv('test.csv',index=None) => index_col=0과 같은 역할
```

```python
!ls
```

```python
df = pd.read_csv('test.csv')
df
```

```python
df = pd.read_csv('test.csv', index_col = 0)
df
```

```python
df1 = df.copy()
df1.to_csv('test1.csv')
df1 = pd.read_csv('test1.csv',index_col=0)
df1
```

```python
df1.iloc[0] = 2
df1.iloc[1] = 3
df1
```

```python
# json 파일 읽고 쓰기
df1.to_json('test2.json')
df_j = pd.read_json('test2.json')
df_j
```

```python
# 바이너리 파일 읽고 쓰기
df1.to_pickle('test3.pkl')
df1 = pd.read_pickle('test3.pkl')
df1
```

```python
# Q. 간단한 데이터프레임 작성 후 csv, json, pickle 형태로 저장 후 다시 불러와서 출력
date = pd.date_range(start = '2024-01-01', periods = 12, freq = 'M')
df1 = pd.DataFrame(date, columns = ['month'])
```

```python
# csv
df1.to_csv('monthly.csv')
df1 = pd.read_csv('monthly.csv', index_col = 0)
df1
```

```python
# json
df1.to_json('monthly.json')
df_j = pd.read_json('monthly.json')
df_j
```

```python
# pikcle
df1.to_pickle('monthly.pkl')
df1_p = pd.read_pickle('monthly.pkl')
df1_p
```
