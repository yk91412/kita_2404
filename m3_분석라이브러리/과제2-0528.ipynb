## 1번

```python
# Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.
import numpy as np
import pandas as pd
df1 = pd.DataFrame(np.arange(12.).reshape((3,4)), columns = list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4,5)), columns = list('abcde'))
```

```python
add = df2.add(df1, fill_value = 0).fillna(0)
sub = df2.sub(df1, fill_value = 0).fillna(0)
mul = df2.mul(df1, fill_value = 0).fillna(0)
div = df2.div(df1, fill_value = 0).fillna(0)

print("데이터 프레임 덧셈:\n",add)
print("데이터 프레임 뺄셈:\n",sub)
print("데이터 프레임 곱셈:\n",mul)
print("데이터 프레임 나눗셈:\n",div)
```

```python
add_data = df1.add(df2, fill_value = 0)
sub_data = df1.sub(df2, fill_value = 0)
mul_data = df1.mul(df2, fill_value = 0)
div_data = df1.div(df2, fill_value = 0)
print("데이터 프레임 덧셈:\n",add_data)
print("데이터 프레임 뺄셈:\n",sub_data)
print("데이터 프레임 곱셈:\n",mul_data)
print("데이터 프레임 나눗셈:\n",div_data)
```

## 3번

```python
# Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])

t1 = df['A'] + s1
t2 = df['B'] + s2
t3 = df['C'] + s3

new_df = pd.concat([t1,t2,t3],axis = 1)
new_df['total'] = new_df.sum(axis=1)
new_df
```

```python
# 여러 열에 시리즈를 더하고 새로운 데이터프레임 생성
df_new = df.copy()
df_new['A'] = df['A'] + s1
df_new['B'] = df['B'] + s2
df_new['C'] = df['C'] + s3

# 각 행의 합계를 계산하여 새로운 열 추가
df_new['Sum'] = df_new.sum(axis = 1)

print(df_new)
```

```python
# Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})
```

```python
t = df['A']+df['B'] > df['C']
t_to_int = t.astype(int)
df['bool'] = t_to_int
df
```

```python
df['A_B_greater_than_C'] = (df['A']+df['B'] > df['C']).astype(int)

df
```
