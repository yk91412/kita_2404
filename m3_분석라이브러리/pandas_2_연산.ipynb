```python
import numpy as np

st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

result = pd.DataFrame([add,sub,mul,div],
                     index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
# Q. nul값을 0으로 변경
# 연산 했을 때 null인 값들을 0으로 변경
result.fillna(0,inplace=True)
result
```

```python
# Q. nul값을 0으로 변경
# null값 자체를 0으로 만들어서 연산
st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})

ad = st1.add(st2,fill_value = 0)
su = st1.sub(st2,fill_value = 0)
mu = st1.mul(st2,fill_value = 0)
di = st1.div(st2,fill_value = 0)

df = pd.DataFrame([ad,su,mu,di],
                  index = ['덧셈','뺄셈','곱셈','나눗셈'])

print(df)
```

```python
# Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])

t1 = df['A'] + s1
t2 = df['B'] + s2
t3 = df['C'] + s3

new_df = pd.concat([t1,t2,t3],axis = 1)
new_df['total'] = new_df.sum(axis=1)
new_df
```

```python
# 여러 열에 시리즈를 더하고 새로운 데이터프레임 생성
df_new = df.copy()
df_new['A'] = df['A'] + s1
df_new['B'] = df['B'] + s2
df_new['C'] = df['C'] + s3

# 각 행의 합계를 계산하여 새로운 열 추가
df_new['Sum'] = df_new.sum(axis = 1)

print(df_new)
```
