```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%A3%BC%EA%B0%84+%EC%98%88%EB%8A%A5+%EC%8B%9C%EC%B2%AD%EB%A5%A0'

# 프로그램 시청률 순위

res = requests.get(url)
html = BeautifulSoup(res.text,'html.parser')
t = html.select('td > p > a')
title = []
channel = []
for i, t in enumerate(t,start = 1):
    if i % 2 == 1:
        title.append(t.text)
    else:
        channel.append(t.text)

for i in range(len(title)):
    print(f"{i+1}위")
    print(f"프로그램명 : {title[i]}")
    print(f"방송사 : {channel[i]}",'\n')
```

```python
url = 'https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query=%EC%A0%9C%EC%A3%BC%EB%8F%84'

res = requests.get(url)

bs = BeautifulSoup(res.text,'html.parser')

titles = bs.select('.title_area')
for title in titles:
    print(title.get_text())
```
