```python
import pandas as pd

df = pd.read_excel('patents_data.xlsx')
df.head()
```

```python
df['application_date'] = df['application_date'].astype(str)
print(df.info())
print(df['application_date'].str[:4])
```

```python
df['year'] = df['application_date'].str[:4]
df
```

```python
def categorize_applicant(x):
    if '현대' in x or 'HYUNDAI' in x:
        return '현대'
    elif '기아' in x or 'KIA' in x:
        return '기아'
    elif '삼성' in x or 'SAMSUNG' in x:
        return '삼성'
    elif '엘지' in x or 'LG' in x:
        return '엘지'
    elif '한화' in x or 'SAMSUNG' in x:
        return '한화'
    elif '애플' in x or 'APPLE' in x:
        return '애플'
    elif '연구원' in x or '연구소' in x:
        return '연구원'
    elif '대학교' in x:
        return '대학교'
    else:
        return '벤처'

df['applicant_cat'] = df['applicant'].apply(lambda x : categorize_applicant(x))
```

```python
topics = {
    '의료분야': ['헬스', '치료', '건강', '유전자', '병원', '환자'],
    '전자상거래': ['상품', '배송', '판매', '결제', '공급망', '판매망'],
    '제너럴 AI': ['추론', '알고리즘', '학습'],
    '자동차': ['주행', '자율주행', '도로', '차량', '교통', '보행자', '차선', '신호등'],
    '금융': ['알고리즘 트레이딩', '리스크', '신용', '금융사기', '자산'],
    '교육': ['교육', '학습 분석', '교육 콘텐츠', '강의', '교육용', '시험', '성적', '학생'],
    '농업': ['스마트 농업', '농작물 모니터링', '농업 로봇', '작물 예측', '농약 사용 최적화', '수확 자동화', '스마트팜'],
    '엔터테인먼트': ['콘텐츠 생성', '게임 AI', '미디어 분석', '사용자 인터페이스', '가상 현실', '음성 인식'],
    '사이버 보안': ['침입 탐지', '데이터 보호', '위협 예측', '인증 시스템', '보안 분석', '암호화 기술']
}

# 주제 정의 함수
def categorize_topic(x):
    for topic, keywords in topics.items():
        if any(kw in x for kw in keywords):
            return topic
    return '기타'  # 주제에 맞는 키워드가 없는 경우

# 'topic' 컬럼에 적용
df['topic'] = df['title'].apply(lambda x: categorize_topic(x))
```
