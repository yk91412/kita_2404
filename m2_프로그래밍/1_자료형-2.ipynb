```python
# 함수에서 여러 값을 반환할 때 튜플을 사용하는 것이 일반적입니다
# 주어진 리스트의 총합과 평균을 동시에 반환하는 함수를 작성하세요
# numbers = [10,20,30,40,50]
def calculate_sum_and_average(numbers):
    add = sum(numbers)
    ave = sum(numbers) / len(numbers)
    return (add,ave)

numbers = [10,20,30,40,50]
# print(calculate_sum_and_average(numbers))
total, average = calculate_sum_and_average(numbers)
# ca..(numbers)에 값이 리턴값으로 값이 두개 저장되어 있으므로 변수를 두개에 저장한다면
# 각 값을 변수에 할당하게 된다 리턴값 순서에 따라  합계는 total에 평균은 average에 저장하게 된다
print(f"리스트에 총합은 {total}이고, 평균은 {average}입니다")
```

```python
from collections import Counter
positions_data = Counter(positions)
print(positions_data)
di1 = dict(positions_data)
print(di1)
print(di1.keys())
```

```python
# 위치 좌표의 리스트
positions = [(1,1), (2,2), (1,1), (3,3), (2,2)]

# 각 위치의 출현 횟수를 저장할 딕셔너리
position_count = {}

for pos in positions :
    if pos in position_count :
        position_count[pos] += 1
    else : #[pos]를 사용하여 인덱싱을 하고 인덱싱 한 곳에 값을 넣는다
        position_count[pos] = 1

print(position_count)
```

```python
# Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

# - numbers = [1, 2, 3, 10, 20]
# - threshold = 15

def max_number(numbers, threshold):
    for i in numbers:
        if i > threshold :
            return True
    return False

numbers = [1,2,3,10]
threshold = 15
print(max_number(numbers,threshold))
# 리스트안에서 그 값보다 하나라도 크다면 참
# return 관계나 논리연산자
```

```python
# Task4_0426. 주어진 문자열이 특정 문자열을 포함하는지 확인하고 결과를 불 값으로 반환하는 함수를 작성하세요.

text = "hello world"
substring = "world"

def string_cheak(text):
    text1 = text.split()
    for i in text1 :
        if i == substring :
            return True
    return False

print(string_cheak(text))
```

```python
# Task5_0426. 주어진 연도가 윤년인지 판별하는 함수를 작성하세요. 윤년은 다음의 조건을 만족해야 합니다:
# - 4로 나누어 떨어진다.
# - 100으로 나누어 떨어지지 않거나, 400으로 나누어 떨어지면 윤년이다.

# year = 2020

# True
def year_check(year):
    if (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0):
        return True
    return False

year = 2022
print(year_check(year))
```

### 사용자 함수

```python
def add(a,b,c):
    return a+b+c

print(add(1,2,3))
```

```python
def mul(a,b,c):
    return a*b*c

print(mul(5,6,7))
```

```python
# 뺄셈 나눗셈을 수행하는 사용자 함수 작성
def multi(a, b):
    minus = a - b
    sksnrl = a / b
    return (minus, sksnrl)

print(multi(10,5))
```

```python
# 뺄셈
def subtract(a,b,c):
    return a - b - c

print(subtract(10,2,2))
```

```python
# 나눗셈
def divide(a,b,c):
    if b == 0 or c == 0:
        return "Error : Division by Zero"
    else :
        return a/b/c

print(round(divide(10,5,9),3))
```

```python
class AddMul:
    def add(self,n1,n2):
        return n1+n2
    def mul(self,n1,n2):
        return n1 * n2

obj1 = AddMul() #객체
print(obj1.add(1,4))
```

```python
# Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요..

class Calculator:
    def add(self,n1,n2):
        return n1 + n2
    def subt(self,n1,n2):
        return n1 - n2
    def mul(self,n1,n2):
        return n1 * n2
    def divi(self,n1,n2):
        return n1 / n2

ob1 = Calculator()
print(ob1.add(5,7))
ob2 = Calculator()
print(ob2.subt(2,7))
ob3 = Calculator()
print(ob3.mul(5,2))
ob4 = Calculator()
print(ob4.divi(9,3))
```

```python
# Task7_0426. 다음 과제를 수행하세요.
# - 사용자로 부터 텍스트를 입력 받는다.
# - 문자을 단어 단위로 분리 : split()
# - 단어의 빈도수를 저장할 딕셔너리를 생성
# - 각 단어의 빈도 수를 계산(for 문 / if else문)
# - 결과 출력

# [ 예시 ]

# 문장을 입력하세요: I love apple. I love orange. Apple is tasty

# {'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}
text = input("텍스트를 입력하세요 : ")
text_lower = text.lower()
text_divide = text_lower.split()

dic = {}
for i in text_divide:
    if i in dic:
        dic[i] += 1
    else :
        dic[i] = 1
print(dic)
```
