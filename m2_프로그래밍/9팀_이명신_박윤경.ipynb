```python
import random
import datetime
class Memo:

    def __init__(self):
        # 운동 리스트
        self.exercise_list = {'1':['숄더 프레스','덤벨 숄더 프레스','사이드 레터럴 레이즈','시티드로우'],
                      '2':['바벨컬','덤벨컬','프레스다운','딥스','덤벨 프레스'],
                      '3':['데드리프트','렛풀다운','바벨로우','덤벨로우','시티드로우'],
                      '4':['스쿼트','런지','레그프레스','레그컬','힙 어브덕션','힙 쓰러스트','힙 브릿지'],
                      '5':['크런치','플랭크','원 레그 레이즈','사이드 크런치','러시안 트위스트'],
                      '6':['덤벨 플라이','체스트 프레스','어시스트 딥','페덱 플라이','인클라인 덤벨 플레스']}

        # 오늘 한 운동 담을 딕셔너리
        self.total = {}

        # 운동 별 칼로리 소모량
        self.calories_burned = {'숄더 프레스': 5, '덤벨 숄더 프레스':5.5, '사이드 레터럴 레이즈':4.5,
                                '시티드로우':6, '바벨컬':5.5, '덤벨컬':5, '프레스다운':6.5, '딥스':7,
                                '덤벨 프레스':5.5, '데드리프트':8, '렛풀다운':6.5, '바벨로우':7.5,
                                '덤벨로우':7, '스쿼트':9,'런지':7.5,'레그프레스':8.5, '레그컬':6,
                                '힙 어브덕션':7, '힙 쓰러스트':8, '크런치':5, '플랭크':4, '원 레그 레이즈':5.5,
                                '사이드 크런치':5, '러시안 트위스트':5.5, '덤벨 플라이':6, '체스트 프레스':7,
                                '어시스트 딥':7.5, '페덱 플라이':6.5, '인클라인 덤벨 플레스':7}


        self.exercise_dic = {'1':'어깨','2':'팔','3':'등','4':'하체','5':'복근','6':'가슴'}

        # 요일 별 운동 해야 할 운동
        self.day = {'Sunday':'휴식', 'Monday': '1, 2, 6', 'Tuesday': '5', 'Wednesday':'3,4',
                    'Thursday':'1, 2, 6', 'Friday':'5', 'Saturday': '3,4'}

        # 음식과 칼로리 값
        self.food_info= {
            '쌀밥':310,'현미밥':321,'김밥':318, '제육볶음':572, '돼지고기김치찌개':128, '불고기':685, '달걀프라이':98,
            '머핀':301,'도너츠':125,'소보루빵':200,'케이크':181, '마늘빵':65,
            '포테이토칩':110,'웨하스':501,'초코하임':537,'새우깡':419,
            '딸기':22,'파인애플':40,'오렌지':50,'포도':48,
            '사이다':105,'콜라':97,'코코아':55,'오렌지쥬스':92,'우유':122}

        self.meal={1:'아침', 2:'점심', 3:'저녁', 4:'야식', 5:'간식'}


        # 먹은 음식 담을 딕셔너리
        self.food_record={}


    # 기초대사량 구하기
    def start_program(self):
        while 1:
            user_weight = int(input("현재 몸무게를 입력해주세요 >> "))
            user_age = int(input("현재 나이를 입력해주세요 >> "))
            user_height = int(input("현재 키를 입력해주세요(cm) >> "))
            user_sex = input("성별을 입력해주세요.(m 또는 f) >> ").strip().lower()
            if user_sex == 'f':
                self.user_bmr = 655 + (9.6 * user_weight) + (1.8 * user_height) - (4.7 * user_age)
                break
            elif user_sex == 'm':
                self.user_bmr = 66 + (13.7 * user_weight) + (5 * user_height) - (6.8 * user_age)
                break
            else:
                print("다시 입력해주세요")
        print("="*40)
        print("하루의 운동량이 어떻게 됩니까?")
        day_exercise = {'1': '주로 앉아서 일하며 운동량이 거의 없다',
                        '2': '약간의 스트레칭 및 몇분이라도 걷는다',
                          '3': '30분이상의 걷기나 운동을 한다',
                          '4': '1시간 이상의 운동을 한다',
                            '5': '많은 땀을 낼 수 있는 고강도의 운동을 한다'}
        for key, value in day_exercise.items():
            print(f"{key}번 {value}")
        day_choice = input("숫자로 입력해주세요 (ex)1 >> ")
        day_cal = {'1':1.2,'2':1.375,'3':1.55,'4':1.725,'5':1.9}
        if day_choice in day_cal:
            for key,value in day_cal.items():
                if key == day_choice:
                    self.user_bmr *= day_cal[key]
                    return self.user_bmr
        else:
            print("다시 입력하세요.")

    # 오늘 기록 선택
    def today_record(self):
        while 1:
            today_record= input("오늘의 식단과 운동을 기록하시겠습니까?(y 또는 n) >> ")
            if today_record == 'y':
                today_record_select = input(f'식단 기록, 운동 기록중 무엇을 기록하시겠습니까? (왼쪽부터 1번입니다) >> ')
                if today_record_select == '1':
                    self.add_meal()
                elif today_record_select=='2':
                    self.add_exercise()
                else:
                    print("없는 번호입니다. 다시 입력해주세요")
            else:
                break

# ============================식단 기록 =============================================
    def meal_category(self):
        for key, value in self.meal.items():
            print(f"번호 : {key}, 식사 시간 : {value}")
        self.meal_choice=input(f'''기록하실 식사를 숫자로 선택해주세요. >>> ''')

    # 해당 식사 때 먹은 음식 입력 받아서 딕셔너리에 저장
    def food_summary(self):
        food=input("\n식사하신 음식을 직접 입력해주세요.\n>>> ")
        if food not in self.food_info:
            add_food_info=input(f"시스텀에 없는 음식입니다. 기록하시려는 {food}의 칼로리를 직접 입력해주세요.")
            self.food_info[food]=add_food_info
        self.food_record[food]=self.food_info[food] # 현재 식사 기록

    # 해당 식사 때 먹은 음식 목록 보여주기
    def total_food_summary(self):
        print(f"식사 목록")
        print("="*40)
        for food, kcal in self.food_record.items():
            print(f"{food} : {kcal}kcal")
        print("="*30)

    # 해당 식사 때 먹은 칼로리 계산
    def total_food_calories(self):
        total_kcal = 0
        for food,kcal in self.food_record.items():
           total_kcal+=int(kcal)
        return total_kcal


    def meal_recommendation(self):
        print(f"음식을 추천해드리겠습니다.")
        food_type_want=int(input(f"어떤 음식을 드시고 싶으십니까?\n1. 일반식\n2. 단백질 중심 \n3. 채식 중심 >> "))
        food_kcal_want=int(input(f"원하시는 최대 칼로리를 입력하세요.\n>>> "))
        self.food_category_1=[['카레라이스',380],['김치볶음밥',430], ['토마토 파스타',550], ['콩국수',320], ['냉면',300], ['탕수육',510]]
        self.food_category_2=[['닭가슴살 웜볼',280],['탄단지 샐러드',340], ['토마호크',420], ['목살 스테이크',280]]
        self.food_category_3=[['지중해식 샐러드',150],['토마토 부라타 샐러드',340], ['연어 포케',290], ['스키야키',510]]


        print(f"음식 추천")
        print("="*40)
        print(f"선호 음식 타입: {food_type_want} \n최대 허용 칼로리: {food_kcal_want}kcal")
        print(f"-"*30)
        if food_type_want==1: # 일반식 선호
            for i in self.food_category_1:
                if (food_kcal_want-50)<=i[1]<=(food_kcal_want+50):
                    print(f'{i[0]} : {i[1]}kcal')

        if food_type_want==2: # 단백질 중심 선호
            for i in self.food_category_2:
                if (food_kcal_want-50)<=i[1]<=(food_kcal_want+50):
                    print(f'{i[0]} : {i[1]}kcal')

        if food_type_want==3: # 채식 중심 선호
            for i in self.food_category_3:
                if (food_kcal_want-50)<=i[1]<=(food_kcal_want+50):
                    print(f'{i[0]} : {i[1]}kcal')

    def add_meal(self):
        today_food = input("식단을 기록하시겠습니까?(y 또는 n) >> ").strip().lower()
        if today_food=='y':
            while True:
                self.meal_category()
                self.food_summary()
                food_add = input("음식 기록을 추가하시겠습니까?(y 또는 n) >> ").strip().lower()
                if food_add=='n':
                    self.total_food_summary()
                    print(f"총 칼로리 섭취량은 {self.total_food_calories()} kcal입니다")
                    if self.user_bmr> self.total_food_calories():
                        print(f"기초대사량은 {self.user_bmr:.2f}로 {self.user_bmr-self.total_food_calories():.2f}kcal 오버 되었습니다.")
                    print(f"오늘의 식단 기록을 종료합니다.")
                    break
                else:
                    continue

# ============================운동 기록 ====================================

    # 저장되어 있는 운동 리스트 보여주기
    def exercise_program(self):
        for key,value in self.exercise_dic.items():
            print(f"번호 : {key}, 운동 부위 : {value}")
        self.today_exercise = input(f'''오늘 운동하신 부위를 숫자로 선택해주세요. >>> ''')

    # 오늘 한 운동 입력 받아서 딕셔너리에 저장
    def exercise_summary(self):
        for key,value in self.exercise_list.items():
            if self.today_exercise == key:
                for i in range(len(value)):
                    machine = input(f"기록할 운동기구를 선택하세요. 기구는 {value}있으며 왼쪽에서 1부터 시작합니다 >> ")
                    exer_set = input("총 몇 세트 진행 하셨습니까? >> ")
                    exer_count = input("한 세트당 몇 번 하셨습니까? >> ")
                    if value[int(machine)-1] not in self.total:
                        self.total[machine]= (value[int(machine)-1],exer_set,exer_count)
                        return self.total

    # 오늘 한 운동 목록 보여주기
    def total_summary(self):
        print("오늘의 운동 목록")
        print("="*40)
        for key,(name,set_count,one_count) in self.total.items():
            print(f"{name} : {one_count}번 {set_count}회")
        print("="*40)

    # 오늘 운동 칼로리 계산
    def exercise_calories(self):
        self.total_calories = 0
        for key, (name, set_count, one_count) in self.total.items():
            calories = self.calories_burned.get(name, 0)
            self.total_calories += calories * int(one_count) * int(set_count)
        return self.total_calories

    # 운동 추천
    def tomorrow_recommend(self):
        current_time = datetime.datetime.now()
        next_time = current_time+datetime.timedelta(days=1)
        next_day = next_time.strftime('%A')
        next_day_exercise = self.day.get(next_day, None)
        if next_day_exercise == '휴식':
            print("내일은 휴식일입니다. 적절한 휴식을 취하세요.")
        else:
            recommend_list = []
            tomorrow_list =[]
            for key in self.exercise_list:
                for i in next_day_exercise:
                    if i== key:
                        tomorrow_list.append(self.exercise_dic[i])
                        recommended_exercises = random.sample(self.exercise_list[key], min(3,len(self.exercise_list[key])))
                        recommend_list.append(recommended_exercises)
            print(f"내일의 운동 부위는 {tomorrow_list}(으)로")
            for i,j in zip(tomorrow_list,recommend_list):
                print(f"추천 운동은 {i} : {j}")

    def add_exercise(self):
            print("="*40)
            print("★ 오늘의 운동 기록하기 ! ★")
            exercise = int(input("오늘 운동한 시간을 분으로 적어주세요 >> "))
            hour, time = divmod(exercise,60)
            print(f"오늘 운동한 시간은 {hour}시간 {time}분 입니다")
            print("="*40)
            while True:
                self.total_summary()
                self.exercise_program()
                self.exercise_summary()
                choice = input("운동 기록을 추가하시겠습니까?(y 또는 n) >> ").strip().lower()
                if choice  == 'n':
                    print("수고하셨습니다 오늘의 운동한 목록을 보여드리겠습니다!")
                    self.total_summary()
                    print(f"오늘의 총 칼로리 소모량은 {self.exercise_calories()} kcal입니다")
                    print(f"기초대사량은 {self.user_bmr:.2f}로 유지를 위해선 {self.user_bmr-self.total_calories:.2f}kcal 소모가 더 필요합니다")
                    self.tomorrow_recommend()
                    break
                else:
                    continue


    # 운동 기록 추가 or 끝마침
    def program(self):
        while True:
            self.start_program()
            self.today_record()
            other_function = input("식단 추천, 운동 추천,프로그램 종료 중 원하시는 번호를 선택해주세요 (왼쪽부터 1번입니다) >> ")
            if other_function == '1':
                self.meal_recommendation()
                other_choice = input("프로그램 종료를 원하십니까?(y 또는 n) >> ").strip().lower()
                if other_choice == 'y':
                    print("프로그램을 종료합니다")
                    break
                else:
                    choice_next = input("내일 할 운동 추천을 원하십니까?(y 또는 n) >> ").strip().lower()
                    if choice_next == 'y':
                        self.tomorrow_recommend()
                        print("내일은 운동을 통해 몸과 마음을 건강하게 유지하세요.")
                        print("더 발전있는 내일을 기원합니다!")
                        break
                    else:
                        choice_num = random.randint(1,6)
                        for key,value in exercise_list.items():
                            if str(choice_num) == key:
                                a = random.sample(value,3)
                                print(f"무작위로 운동 추천을 해드립니다.부위는 {self.exercise_dic[key]}이며 {a}를 추천해드립니다")
                                print("내일은 운동을 통해 몸과 마음을 건강하게 유지하세요.")
                                print("더 발전있는 내일을 기원합니다!")
                                break
            elif other_function == '2':
                    choice_next = input("내일 할 운동 추천을 원하십니까?(y 또는 n) >> ").strip().lower()
                    if choice_next == 'y':
                        self.tomorrow_recommend()
                        print("내일은 운동을 통해 몸과 마음을 건강하게 유지하세요.")
                        print("더 발전있는 내일을 기원합니다!")
                        other_choice = input("프로그램 종료를 원하십니까?(y 또는 n) >> ").strip().lower()
                        if other_choice == 'y':
                            print("프로그램을 종료합니다")
                            break
                        else:
                            choice_next = input("식단 추천을 원하십니까?(y 또는 n) >> ").strip().lower()
                            if choice_next == 'y':
                                self.meal_recommendation()
                                print("프로그램을 종료하겠습니다")
                                break
                            else:
                                print("프로그램을 종료하겠습니다")
                    else:
                        choice_num = random.randint(1,6)
                        for key,value in exercise_list.items():
                            if str(choice_num) == key:
                                a = random.sample(value,3)
                                print(f"무작위로 운동 추천을 해드립니다.부위는 {self.exercise_dic[key]}이며 {a}를 추천해드립니다")
                                print("내일은 운동을 통해 몸과 마음을 건강하게 유지하세요.")
                                print("더 발전있는 내일을 기원합니다!")
                                choice_next = input("식단 추천을 원하십니까?(y 또는 n) >> ").strip().lower()
                                if choice_next == 'y':
                                    self.meal_recommendation()
                                    print("프로그램을 종료하겠습니다")
                                    break
                                else:
                                    print("프로그램을 종료하겠습니다")
                                    break
            else:
                print("프로그램을 종료합니다")
                break
            break
if __name__ == "__main__":
    exercise_memo = Memo()
    exercise_memo.program()

```
