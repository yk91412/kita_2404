```python
# 사용자 함수 활용
def cal_factorial(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result

print(cal_factorial(5))
print(cal_factorial(10))
```

```python
def math(x,y,z):
    if z == '+':
        return x+y
    elif z == '-':
        return x-y
    elif z == '*':
        return x*y
    elif z == '/':
        return x/y
    elif z == '//':
        return x//y
    elif z == '%':
        return x%y
    elif z=='^':
        return x^y
    else:
        return

print(math(4,6,'+'))
print(math(int(input("숫자1 : ")), int(input("숫자 2 : ")), input("기호 : ")))
```

```python
# Task10_0509. 위의 계산을 연산자를 choice로 가변적인 숫자들을 *args로 사용한 사용자 함수 cal을 이용하여 다음을 수행하세요.

# cal('*',1,2,3,4,5)
# cal('+',1,2,3,4,5,6,7)
# cal('-',1,2,3,4,5,6,7)
# cal('/',10,3)-> 0일수도 있으니 에러 처리함께
def cal(choice,*args):
    if choice == '+':
        print(sum(args))
    elif choice == '-':
        total_minus = 0
        for i in args:
            total_minus -= i
        print(total_minus)
    elif choice == '*':
        total_mul = 1
        for i in args:
            total_mul *= i
        print(total_mul)
    elif choice == '/':
        total_divide = args[0]
        for i in args[1:]:
            if i != 0:
                total_divide /= i
                print(total_divide)
            else:
                print("0으로 나눌 수 없습니다")
    else:
        print("연산 할 수 없는 연산자 입니다")

cal('/',10,4)
```
